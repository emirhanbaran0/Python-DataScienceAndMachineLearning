# Çaprazlamalar

```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds=sns.load_dataset("diamonds")
df=diamonds.copy()
cut_categoriler=["Fair","Good","Very Good","Premium", "Ideal"]
df.cut=df.cut.astype(CategoricalDtype(ordered=True,categories=cut_categoriler))
df.head()
```

```python
sns.catplot(x="cut" , y="price" , data=df)
```

```python
# Yorum yapacak olursak price değerinin arttığı yelerde cut değişkenine ait 
# üst sınıflar daha çok görülmektedir.
# Örneğin Ideal sınıfı yukarıda çok yoğunken fair sınıfı 8-9 bin aşağısında i
# çok yoğundur.
```

```python
sns.barplot(x="cut",y="price",hue="color",data=df)
```

```python
# Fair, good ,... sınıflarının kendi içinde color'a göre nasl yapılandığını gözlemleyebildik.
```

```python
df.groupby(["cut","color"])["price"].mean()
```

```python
df.groupby(["cut","clarity"])["price"].mean()
```

```python
sns.barplot(x="cut",y="price",hue="clarity",data=df)
```
