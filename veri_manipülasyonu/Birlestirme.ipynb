# Birleştirme(Join) İşlemleri

```python
import numpy as np 
import pandas as pd
m=np.random.randint(1,30,size=(5,3))
df1=pd.DataFrame(m,columns=["var1","var2","var3"])
df1
```

```python
df2=df1+122
df2
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
?pd.concat
```

```python
df2.columns=["var1","var2","deg3"]
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],join="inner",ignore_index=True)
```

```python
?pd.concat
```

# İleri Birleştirme İşlemleri

```python
#birebir birleştirme
```

```python
df1=pd.DataFrame({'calisanlar': ["Ali","Veli","Ayşe","Fatma"],
                   'grup': ["Muhasebe","Mühendislik","Mühendislik","İK"]})
df1
```

```python
df2=pd.DataFrame({'calisanlar': ["Ayşe","Ali","Veli","Fatma"],
                   'giris': [2010,2009,2014,2019]})
df2
```

```python
pd.merge(df1,df2,on="calisanlar")
```

```python
# many to one
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df4=pd.DataFrame({'grup': ["Muhasebe","Mühendislik","İK"],
                   'mudur': ['Caner','Mustafa','Berkcan']})
df4
```

```python
pd.merge(df4,df3)
```
